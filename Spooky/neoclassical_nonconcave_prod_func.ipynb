```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fontsize= 14
ticksize = 14
figsize = (8,5)
params = {'font.family':'serif',
    "figure.figsize":figsize,
    'figure.dpi': 80,
    'figure.edgecolor': 'k',
    'font.size': fontsize,
    'axes.labelsize': fontsize,
    'axes.titlesize': fontsize,
    'xtick.labelsize': ticksize,
    'ytick.labelsize': ticksize
}
plt.rcParams.update(params)
```

```python
def f(k):
    a = 0.5
    b_1 = 3.0
    b_2 = 2.5
    α = 0.333
    term1 = k**α
    term2 = b_1 * (k**α) - b_2
    return a*term1, a*term2, a*np.maximum(term1, term2)
```

```python
x = np.linspace(0.1,5,500)
prod_first, prod_sec, prod = f(x)
```

```python
plt.plot(x,prod_first, linestyle = ':', color = 'red')
plt.plot(x,prod_sec, linestyle = ':', color = 'blue')
plt.plot(x,prod, color = 'k', label = r"Production function")
plt.ylim([0,1.75])
plt.title(r"$f(k) = a~\max\{k^\alpha,b_1 k^\alpha-b_2\}  $")
plt.xlabel(r"Capital ($k$)")
plt.ylabel(r"Production")
plt.xticks([])
plt.yticks([])
plt.legend()
plt.savefig('non_concave_prod_func.pdf')
```
