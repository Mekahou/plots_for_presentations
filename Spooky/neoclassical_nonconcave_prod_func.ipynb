```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fontsize= 14
ticksize = 14
figsize = (10,5)
params = {'font.family':'serif',
    "figure.figsize":figsize,
    'figure.dpi': 80,
    'figure.edgecolor': 'k',
    'font.size': fontsize,
    'axes.labelsize': fontsize,
    'axes.titlesize': fontsize,
    'xtick.labelsize': ticksize,
    'ytick.labelsize': ticksize
}
plt.rcParams.update(params)
```

```python
def f(k):
    a = 0.5
    b_1 = 3.0
    b_2 = 2.5
    α = 0.333
    term1 = k**α
    term2 = b_1 * (k**α) - b_2
    return a*np.maximum(term1, term2)
```

```python
x = np.linspace(0,5,500)
y = f(x)
```

```python
plt.plot(x,y)
```
